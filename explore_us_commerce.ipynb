```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data 
dat = pd.read_csv('US  E-commerce records 2020.csv', encoding= 'unicode_escape' )
```

```python
# Sort the dataframe by the Name column in alphabetical order
dat_sorted  = dat.sort_values(by='State', ascending=False)

# Plot size
plt.figure(figsize = (20,10),dpi = 100)

x = dat_sorted['Sales']
y = dat_sorted['State']

# Create the barplot
plt.barh(y, x, align='center')

# Add labels and a title
plt.xlabel('Sales')
plt.ylabel('State')
plt.title('Sales by State')

# Adjust the plot layout
plt.tight_layout()
```
